```python
def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 10)

def check_win(board, player):
    win_conditions = [
        [board[0][0], board[0][1], board[0][2]],  # row 1
        [board[1][0], board[1][1], board[1][2]],  # row 2
        [board[2][0], board[2][1], board[2][2]],  # row 3
        [board[0][0], board[1][0], board[2][0]],  # column 1
        [board[0][1], board[1][1], board[2][1]],  # column 2
        [board[0][2], board[1][2], board[2][2]],  # column 3
        [board[0][0], board[1][1], board[2][2]],  # diagonal 1
        [board[2][0], board[1][1], board[0][2]]   # diagonal 2
    ]
    return [player, player, player] in win_conditions

def get_free_positions(board):
    return [(r, c) for r in range(3) for c in range(3) if board[r][c] == " "]

def main():
    board = [[" " for _ in range(3)] for _ in range(3)]
    current_player = "X"
    
    while True:
        print_board(board)
        free_positions = get_free_positions(board)
        if not free_positions:
            print("It's a draw!")
            break

        print(f"Player {current_player}'s turn.")
        try:
            row, col = map(int, input("Enter row and column numbers to place your mark (0-2): ").split())
            if (row, col) in free_positions:
                board[row][col] = current_player
                if check_win(board, current_player):
                    print_board(board)
                    print(f"Player {current_player} wins!")
                    break
                current_player = "O" if current_player == "X" else "X"
            else:
                print("This position is already taken or out of bounds, try again.")
        except ValueError:
            print("Invalid input. Please enter two numbers separated by space.")

if __name__ == "__main__":
    main()
```
